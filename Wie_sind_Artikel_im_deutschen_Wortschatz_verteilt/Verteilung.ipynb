```python
counts = pd.DataFrame()

with pd.read_csv("nouns.csv", chunksize=5000) as reader:
    for chunk in reader:
        counts = counts.append(chunk["genus"].value_counts())
        
counts
```

```python
df = pd.DataFrame(counts.sum())[0]
df.plot.pie(labels=["Feminin", "Maskulin", "Neutral"], autopct='%1.1f%%', startangle=90, figsize=(15, 15), fontsize=20)
```
